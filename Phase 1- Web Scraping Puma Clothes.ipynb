```python
def scroll_to_loading():
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(3)
    driver.execute_script("window.scrollBy(0, -400);")


def fetch_product_urls():
    product_urls = []
    product_elements = driver.find_elements(By.XPATH, '//li[@data-test-id="product-list-item"]/a')
    for product_element in product_elements:
        product_urls.append(product_element.get_attribute('href'))
    return product_urls


product_urls = []


while len(product_urls) < 5000:
    
    scroll_to_loading()
    
    time.sleep(2)
    
    new_product_urls = fetch_product_urls()
    product_urls.extend(new_product_urls)


for product_url in product_urls:
    print(product_url)
```
